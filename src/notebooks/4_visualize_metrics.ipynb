```python
import pandas as pd
import plotly.express as px
```

```python
experiments = ["ollama_qwen3_4b_parallel", "ollama_qwen3_4b_serial", "vllm_qwen3_4b_parallel"]
data = {
    experiment: pd.read_csv(f"../data/{experiment}/hardware_metrics_log.csv", index_col="timestamp_utc")
    for experiment in experiments
}
```

```python
experiment = "vllm_qwen3_4b_parallel"

px.line(
    data[experiment],
    y=["cpu_percent", "gpu_mem_percent", "gpu_utilization_percent", "vmem_percent"],
    title="System Resource Usage Over Time (vLLM Inference - Parallel Requests)",
    labels={"value": "Usage Percentage", "timestamp_utc": "Timestamp (UTC)"},
).show()
```
