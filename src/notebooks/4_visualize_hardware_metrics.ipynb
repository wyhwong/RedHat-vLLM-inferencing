```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv(f"../data/vllm_testing/hardware_metrics_log.csv", index_col="timestamp_utc")
```

```python
px.line(
    data,
    y=["cpu_percent", "gpu_mem_percent", "gpu_utilization_percent", "vmem_percent"],
    title="System Resource Usage Over Time",
    labels={"value": "Usage Percentage", "timestamp_utc": "Timestamp (UTC)"},
    width=1200,
    height=600,
).show()
```
