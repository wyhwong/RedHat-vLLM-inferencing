## Sample 50 data points from Alpaca dataset

```python
import pandas as pd
```

```python
SAMPLE_OUTPUT_FILEPATH = "../data/50_data_points_from_alpaca.csv"
```

```python
df = pd.read_parquet(
    "hf://datasets/tatsu-lab/alpaca/data/train-00000-of-00001-a09b74b3ef9c3b56.parquet"
)
# Here we only want data points without any input for convenience
df = df.loc[df["input"] == ""].sample(50, replace=False, random_state=2026).drop(columns=["input"])
df.to_csv(SAMPLE_OUTPUT_FILEPATH, index=False)
```
